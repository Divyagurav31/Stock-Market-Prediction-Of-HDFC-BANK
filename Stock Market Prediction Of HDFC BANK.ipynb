```python
from sklearn.svm import SVR  # Support Vector Regressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
svr = SVR()
```

```python
def mymodel(model):
    model.fit(xtrain, ytrain)  # Call the fit method directly on the model object
    ypred = model.predict(xtest)
    mse = mean_squared_error(ytest, ypred)
    mae = mean_absolute_error(ytest, ypred)
    r2 = r2_score(ytest, ypred)
    print(f'Mean Squared Error: {mse}')
    print(f'Mean Absolute Error: {mae}')
    print(f'R^2 Score: {r2}')
```

```python
# Call the function with the SVR model
mymodel(svr)
```

```python
def mymodel1(model, xtrain_sc, ytrain, xtest_sc, ytest):
    model.fit(xtrain_sc, ytrain)  # Use the model argument directly
    ypred = model.predict(xtest_sc)
    mse = mean_squared_error(ytest, ypred)
    mae = mean_absolute_error(ytest, ypred)
    r2 = r2_score(ytest, ypred)
    return mse, r2

# Example usage with SVR
from sklearn.svm import SVR

svr = SVR()
mse, r2 = mymodel1(svr, xtrain_sc, ytrain, xtest_sc, ytest)
print(f'Mean Absolute Error: {mae}')
print(f"Mean Squared Error: {mse}")
print(f"R^2 Score: {r2}")
```

```python
#Hypertuning and Pipeline
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
p1=Pipeline(steps=[("scaling",StandardScaler()),('svr',SVR()),])
```

```python
p1.fit(xtrain,ytrain)
ypred=p1.predict(xtest) 
mse = mean_squared_error(ytest, ypred)
mae = mean_absolute_error(ytest, ypred)
r2 = r2_score(ytest, ypred)
print(f'Mean Squared Error: {mse}')
print(f'Mean Absolute Error: {mae}')
print(f'R^2 Score: {r2}')
```

<h4 style="color: #00005C;border:3px dotted black; padding:10px;margin:0 200px ">
        <b>Comparing All Algorithms</b></h4>
</div>

<p>To compare the performance of all algorithms used in the project, you can create a summary of the evaluation metrics (MSE, MAE, and R² score) for each model. This will help in identifying the best-performing model.</p>

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import pandas as pd

# Define a function to evaluate and return the metrics
def evaluate_model(model, xtrain, ytrain, xtest, ytest):
    model.fit(xtrain, ytrain)
    ypred = model.predict(xtest)
    mse = mean_squared_error(ytest, ypred)
    mae = mean_absolute_error(ytest, ypred)
    r2 = r2_score(ytest, ypred)
    return mse, mae, r2
```

```python
# Models to evaluate
models = {
    "Multilinear Regression": multi_reg,
    "SVR": svr,
    "Gradient Boosting": gb_model,
    "Random Forest": rf_regressor1
}
```

```python
# Collect metrics for each model
metrics = []

for model_name, model in models.items():
    mse, mae, r2 = evaluate_model(model, xtrain_sc, ytrain, xtest_sc, ytest)
    metrics.append({
        "Model": model_name,
        "MSE": mse,
        "MAE": mae,
        "R² Score": r2
    })
```

```python
# Convert metrics to a DataFrame for better visualization
metrics_df = pd.DataFrame(metrics)
print(metrics_df)
```
